```python
cv2.rectangle(m,(55,30),(85,60),(0,0,0),-1)
plt.imshow(m, cmap='gray', vmin=0, vmax=255)
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
mWithLaplacianFilter = cv2.Laplacian(m, cv2.CV_64F)

minLoG = cv2.morphologyEx(mWithLaplacianFilter, cv2.MORPH_ERODE, np.ones((3,3)))
maxLoG = cv2.morphologyEx(mWithLaplacianFilter, cv2.MORPH_DILATE, np.ones((3,3)))
zeroCrossM = np.logical_or(np.logical_and(minLoG < 0, mWithLaplacianFilter > 0), np.logical_and(maxLoG > 0, mWithLaplacianFilter < 0)) 

plt.imshow(zeroCrossM, cmap='gray')
plt.title('Zero-Crossings')
plt.xticks([]), plt.yticks([])
plt.show()
```
