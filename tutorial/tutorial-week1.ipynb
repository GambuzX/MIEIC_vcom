```python
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt

dataDir = './data'
```

```python
# Opening an image
img = cv2.imread(os.path.join(dataDir, 'ml.png'))
plt.imshow(img)
plt.show()
```

```python
# Matplotlib expects RGB format, but OpenCV provides it in BGR
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
plt.imshow(img)
plt.show()
```

```python
# Create a black image
m = np.zeros((512,512,3), np.uint8)
plt.imshow(m)
plt.show()
```

```python
# Draw a diagonal blue line with thickness of 5 px
cv2.line(m,(0,0),(511,511),(0,0,255),5)
plt.imshow(m)
plt.show()
```

```python
# Draw another diagonal blue line with thickness of 5 px
cv2.line(m,(0,511),(511,0),(0,0,255),5)
plt.imshow(m)
plt.show()
```

```python
# Draw a green square on the top right-most corner with line with thickness of 3 px
cv2.rectangle(m,(384,0),(510,128),(0,255,0),3)
plt.imshow(m)
plt.show()
```

```python
bgColor = (255,0,0)
dimensions = (512,512,3)
m2 = np.full(dimensions, bgColor, np.uint8) # draw background

squareSize = 100
squareX = (int(dimensions[0]/2)-squareSize, int(dimensions[1]/2)-squareSize)
squareY = (int(dimensions[0]/2)+squareSize, int(dimensions[1]/2)+squareSize)
squareColor = (255,255,255)
squareThickness = 3

circleRadius = 100
circleCenter = (int(dimensions[0]/2), int(dimensions[1]/2))
circleColor = (255,255,255)
circleThickness = 3

cv2.rectangle(m2, squareX, squareY, squareColor, squareThickness)
cv2.circle(m2, circleCenter, circleRadius, circleColor, circleThickness)

plt.imshow(m2)
plt.savefig("data/square_circle.png")
plt.show()
```

```python
img2 = cv2.imread(os.path.join(dataDir, 'ml.png'))

cv2.rectangle(img2, (50,50), (155, 210), (255,0,0), 3)
plt.imshow(img2)
plt.show()
```
